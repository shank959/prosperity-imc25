```python
def prize(multiplier, inhabitants, percentage):
    return (10_000 * multiplier)/(inhabitants + percentage)

containers = [(10, 1),
              (80, 6),
              (37, 3),
              (17, 1),
              (31, 2),
              (90, 10),
              (50, 4),
              (20, 2),
              (74, 4),
              (89, 8)]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
potential_prizes = [[prize(multiplier, inhabitants, percentage) for percentage in np.arange(0, 100.1, 1)] for multiplier, inhabitants in containers]

plt.figure(figsize=(10, 6))
plt.grid()
plt.title('Potential Prizes')
plt.xlabel('Percentage')
plt.ylabel('Potential Prize')
for idx, (multiplier, inhabitants) in enumerate(containers):
    plt.plot(np.arange(0, 100.1, 1), potential_prizes[idx], label=f'Multiplier {multiplier}')
plt.legend()
```
